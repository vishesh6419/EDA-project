```python
def get_mean_median_by_network(df, network_name):
    # Filter for calls made to a specific network
    data = df[(df['item'] == 'call') & (df['network'] == network_name)]

    if data.empty:
        print("Invalid network name or no call records found.")
        return

    print("Mean values:\n", data.mean(numeric_only=True))
    print("\nMedian values:\n", data.median(numeric_only=True))
```

```python
# Step 1: Filter only call activity (similar to canceled bookings)
df_calls = df[df['item'] == 'call']

# Step 2: Calls made to a specific network, e.g., Vodafone (like 'No Deposit')
df_calls_vodafone = df_calls[df_calls['network'] == 'Vodafone']

# Step 3: Calls made to other networks (like having a deposit)
df_calls_other_networks = df_calls[df_calls['network'] != 'Vodafone']

# Step 4: Calls made during peak hours (e.g., 9 AM to 6 PM)
df_calls_peak_hours = df_calls[(df_calls['hour'] >= 9) & (df_calls['hour'] <= 18)]

# Step 5: Calls made during off-peak hours
df_calls_off_peak = df_calls[(df_calls['hour'] < 9) | (df_calls['hour'] > 18)]

# Step 6: Long-duration calls (e.g., over 60 seconds)
df_calls_long = df_calls[df_calls['duration'] > 60]

# Step 7: Short-duration calls (60 seconds or less)
df_calls_short = df_calls[df_calls['duration'] <= 60]
```

```python
# 1. Make a working copy
df = df.copy()

# 2. Convert 'date' to datetime (if not already done)
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y %H:%M')

# 3. Drop unnecessary columns (like index if present)
if 'index' in df.columns:
    df.drop(columns=['index'], inplace=True)

# 4. Handle missing or invalid values
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')  # ensures numeric duration
df.dropna(subset=['duration'], inplace=True)  # remove rows with invalid duration if any

# 5. Create time-based features
df['hour'] = df['date'].dt.hour
df['day'] = df['date'].dt.day_name()
df['month_num'] = df['date'].dt.month
df['weekday'] = df['date'].dt.weekday
df['week'] = df['date'].dt.isocalendar().week

# 6. Standardize categorical values
df['item'] = df['item'].str.lower().str.strip()
df['network'] = df['network'].str.strip()
df['network_type'] = df['network_type'].str.lower().str.strip()

# 7. Categorize duration for deeper insights
df['duration_category'] = pd.cut(df['duration'],
                                 bins=[0, 30, 60, 180, np.inf],
                                 labels=['Very Short', 'Short', 'Moderate', 'Long'])

# 8. Create 'time_of_day' buckets (optional but insightful)
def map_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['time_of_day'] = df['hour'].apply(map_time_of_day)

# 9. Remove outliers (optional but useful for clean visuals)
q99 = df['duration'].quantile(0.99)
df = df[df['duration'] <= q99]

# 10. Reset index after wrangling
df.reset_index(drop=True, inplace=True)

# Final check
print("✅ Data wrangling complete. Final shape:", df.shape)
df.head()
```

In the data wrangling process, I thoroughly cleaned and enriched your phone usage dataset to prepare it for meaningful analysis. I started by converting the date column into a proper datetime format so we could easily extract time-based features like hour, weekday, and month. This allowed us to break down usage trends across different times of day and days of the week. I also ensured the call duration values were numeric and removed any records with invalid or missing durations to maintain data quality. To make comparisons more accurate, I standardized the text in the item, network, and network_type columns by converting them to lowercase and trimming extra spaces. For better segmentation, I categorized call durations into labels like 'Very Short', 'Short', 'Moderate', and 'Long', and also created custom time-of-day buckets such as 'Morning', 'Afternoon', 'Evening', and 'Night'. Additionally, I handled outliers by removing extremely long-duration calls (above the 99th percentile) to avoid skewing insights. After all transformations, I reset the index to keep the dataset clean and ready for further analysis.

From these manipulations, we uncovered several useful insights. We found that the majority of the activity in the dataset consists of calls, with Vodafone being one of the most used networks. Users tend to make more calls during the daytime, especially between 9 AM and 6 PM, while night hours show a noticeable drop in activity. #### Chart - 2

```python
# Chart - 2 visualization code
import matplotlib.pyplot as plt
import numpy as np

# Set figure size
plt.rcParams['figure.figsize'] = (12, 7)

# Filter only calls (exclude SMS/data)
df_calls = df[df['item'] == 'call']

# Step 1: Group by network and calculate total calls & long call %
network_stats = df_calls.groupby('network').apply(
    lambda x: pd.Series({
        'Total Calls': len(x),
        'Long Call %': (x['duration'] > 60).mean() * 100
    })
)

# Step 2: Filter networks with at least 50 calls
filtered_stats = network_stats[network_stats['Total Calls'] >= 50]

# Step 3: Sort and select top 10 networks with highest long call percentage
top10_filtered = filtered_stats.sort_values('Long Call %', ascending=False).head(10)

# Step 4: Print the result
print("Top 10 Networks with Highest Long Call Percentage (Min 50 calls):")
print(top10_filtered.reset_index())
print(" ")

# Step 5: Define color palette
color = plt.cm.viridis(np.linspace(0, 0.7, 10))

# Step 6: Plot bar chart
top10_filtered['Long Call %'].plot(kind='bar', color=color)
plt.title("Top 10 Networks with Highest Long Call Percentage (Min 50 Calls)", fontsize=16)
plt.xlabel("Network", fontsize=14)
plt.ylabel("Long Call Percentage (%)", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
*I chose a bar chart for this analysis because it clearly compares the long call percentage across different networks, making it easy to see which ones stand out. #### Chart - 14 - Correlation Heatmap

```python
# Correlation Heatmap visualization code
df['date'] = pd.to_datetime(df['date'])

# Extract additional numerical features from date
df['day_of_week'] = df['date'].dt.dayofweek  # Monday=0
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

# Select only numerical columns for correlation
num_cols = df[['duration', 'day_of_week', 'month', 'day']]

# Calculate correlation matrix
corr = num_cols.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Phone Usage Features')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?
A heatmap is ideal for visualizing correlations between multiple numeric variables. It gives a bird’s-eye view of how strongly (or weakly) variables are related, using color intensity and numerical values.

positive correlation is observed, it suggests that phone usage increases in certain months — indicating possible seasonal trends (like school terms or holidays).

#### Chart - 15 - Pair Plot

```python
# Pair Plot visualization code
df['date'] = pd.to_datetime(df['date'])

# Create additional time-based numeric columns
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek

# Select numerical features for pair plot
pairplot_data = df[['duration', 'month', 'day', 'day_of_week', 'item']]

# Create pair plot with hue as 'item' (call, sms, data)
sns.pairplot(pairplot_data, hue='item', palette='Set2', corner=True)
plt.suptitle("Pair Plot of Phone Usage Features by Item Type", y=1.02)
plt.show()
```
